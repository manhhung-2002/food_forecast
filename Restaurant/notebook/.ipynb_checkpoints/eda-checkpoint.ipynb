```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
import matplotlib.pyplot as plt
import seaborn as sns

# Kh·ªüi t·∫°o Spark
spark = SparkSession.builder.appName("RestaurantEDA").getOrCreate()

# Load c√°c b·∫£ng CSV
tables = ["dish", "order", "order_detail"]
dfs = {
    t: spark.read.csv(f"../data/{t}.csv", header=True, inferSchema=True)
    for t in tables
}

# ƒê·ªïi t√™n c·ªôt time ‚Üí order_time cho b·∫£ng order
dfs["order"] = dfs["order"].withColumnRenamed("time", "order_time")

# √âp ki·ªÉu timestamp
dfs["order"] = dfs["order"].withColumn("order_time", F.to_timestamp("order_time"))

# Ki·ªÉm tra schema
for name, df in dfs.items():
    print(f"\nüìÇ Schema {name}")
    df.printSchema()
    print(f"üî¢ S·ªë d√≤ng: {df.count()}")

```

```python
# T√≠nh s·ªë order theo qu√Ω
orders_per_quarter = (
    dfs["order"]
    .withColumn("year", F.year("order_time"))
    .withColumn("quarter", F.quarter("order_time"))
    .groupBy("year", "quarter")
    .count()
    .orderBy("year", "quarter")
    .toPandas()
)


# V·∫Ω bi·ªÉu ƒë·ªì
plt.figure(figsize=(8, 5))
sns.barplot(data=orders_per_quarter, x="quarter", y="count", hue="year", palette="Set2")

plt.title("üì¶ S·ªë l∆∞·ª£ng ƒë∆°n h√†ng theo qu√Ω")
plt.xlabel("Qu√Ω")
plt.ylabel("S·ªë ƒë∆°n h√†ng")
plt.legend(title="NƒÉm")
plt.show()
```

Trong nƒÉm 2024, s·ªë l∆∞·ª£ng ƒë∆°n h√†ng c√≥ s·ª± bi·∫øn ƒë·ªông r√µ r·ªát theo t·ª´ng qu√Ω. Qu√Ω 1 ghi nh·∫≠n m·ª©c th·∫•p nh·∫•t v·ªõi kho·∫£ng 9.000 ƒë∆°n h√†ng. Sang Qu√Ω 2, s·ªë ƒë∆°n tƒÉng g·∫•p ƒë√¥i l√™n g·∫ßn 18.000 ƒë∆°n v√† ƒë·∫°t ƒë·ªânh ·ªü Qu√Ω 3 v·ªõi kho·∫£ng 20.000 ƒë∆°n. Tuy nhi√™n, ƒë·∫øn Qu√Ω 4, s·ªë l∆∞·ª£ng ƒë∆°n h√†ng gi·∫£m c√≤n kho·∫£ng 14.000, nh∆∞ng v·∫´n cao h∆°n so v·ªõi Qu√Ω 1. Nh√¨n chung, xu h∆∞·ªõng th·ªÉ hi·ªán s·ª± tƒÉng tr∆∞·ªüng m·∫°nh m·∫Ω t·ª´ Qu√Ω 1 ƒë·∫øn Qu√Ω 3 tr∆∞·ªõc khi ch·ªØng l·∫°i v√† gi·∫£m nh·∫π v√†o Qu√Ω 4.


```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

import matplotlib
matplotlib.rcParams['font.family'] = 'DejaVu Sans'  # font h·ªó tr·ª£ unicode

import matplotlib.pyplot as plt
import seaborn as sns
import pyspark.sql.functions as F

sns.set(style="whitegrid")

# T√≠nh s·ªë order theo th√°ng
orders_per_month = (
    dfs["order"]
    .withColumn("year", F.year("order_time"))
    .withColumn("month", F.month("order_time"))
    .groupBy("year", "month")
    .count()
    .orderBy("year", "month")
    .toPandas()
)

# 2Ô∏è‚É£ Bar chart: so s√°nh th√°ng gi·ªØa c√°c nƒÉm
plt.figure(figsize=(12, 6))
sns.barplot(
    data=orders_per_month,
    x="month",
    y="count",
    hue="year",
    palette="Set2"
)
plt.title("üì¶ S·ªë l∆∞·ª£ng ƒë∆°n h√†ng theo t·ª´ng th√°ng (Bar Chart)")
plt.xlabel("Th√°ng")
plt.ylabel("S·ªë ƒë∆°n h√†ng")
plt.xticks(range(1, 13))
plt.legend(title="NƒÉm")
plt.show()

```

Trong nƒÉm 2024, s·ªë l∆∞·ª£ng ƒë∆°n h√†ng ph√¢n b·ªï theo th√°ng c√≥ s·ª± thay ƒë·ªïi ƒë√°ng k·ªÉ. T·ª´ th√°ng 1 ƒë·∫øn th√°ng 4, l∆∞·ª£ng ƒë∆°n h√†ng duy tr√¨ ·ªü m·ª©c th·∫•p, dao ƒë·ªông t·ª´ kho·∫£ng 2.700 ƒë·∫øn 3.200 ƒë∆°n. B·∫Øt ƒë·∫ßu t·ª´ th√°ng 5, ƒë∆°n h√†ng tƒÉng v·ªçt l√™n h∆°n 7.500 ƒë∆°n v√† gi·ªØ ·ªü m·ª©c cao trong c√°c th√°ng 5 ƒë·∫øn 8, trong ƒë√≥ th√°ng 8 ƒë·∫°t ƒë·ªânh g·∫ßn 8.000 ƒë∆°n. T·ª´ th√°ng 9 tr·ªü ƒëi, s·ªë ƒë∆°n h√†ng gi·∫£m d·∫ßn, dao ƒë·ªông quanh m·ª©c 4.300 ‚Äì 5.000 ƒë∆°n, th·∫•p h∆°n so v·ªõi giai ƒëo·∫°n cao ƒëi·ªÉm gi·ªØa nƒÉm nh∆∞ng v·∫´n cao h∆°n giai ƒëo·∫°n ƒë·∫ßu nƒÉm. Nh√¨n chung, xu h∆∞·ªõng th·ªÉ hi·ªán s·ª± b√πng n·ªï ƒë∆°n h√†ng v√†o m√πa gi·ªØa nƒÉm, sau ƒë√≥ gi·∫£m v·ªÅ cu·ªëi nƒÉm.

```python
# Doanh thu theo qu√Ω
revenue_quarter = (
    order_detail_with_time
    .withColumn("quarter", F.quarter("order_time"))
    .groupBy("quarter")
    .agg((F.sum(F.col("quantity") * F.col("price"))).alias("total_revenue"))
    .orderBy("quarter")
    .toPandas()
)

plt.figure(figsize=(7, 5))
sns.barplot(data=revenue_quarter, x="quarter", y="total_revenue", palette="Blues_d")
plt.title("üìÖ Doanh thu theo qu√Ω")
plt.xlabel("Qu√Ω")
plt.ylabel("Doanh thu")
plt.show()
```

Doanh thu nƒÉm 2024 c√≥ s·ª± tƒÉng tr∆∞·ªüng r√µ r·ªát trong ba qu√Ω ƒë·∫ßu. Qu√Ω 1 ghi nh·∫≠n m·ª©c th·∫•p nh·∫•t v·ªõi kho·∫£ng 6.000 t·ª∑ ƒë·ªìng. Sang Qu√Ω 2, doanh thu tƒÉng g·∫•p ƒë√¥i l√™n g·∫ßn 12.000 t·ª∑ v√† ƒë·∫°t ƒë·ªânh ·ªü Qu√Ω 3 v·ªõi kho·∫£ng 13.500 t·ª∑ ƒë·ªìng. Tuy nhi√™n, ƒë·∫øn Qu√Ω 4, doanh thu gi·∫£m xu·ªëng c√≤n kho·∫£ng 10.000 t·ª∑, th·∫•p h∆°n Qu√Ω 2 v√† Qu√Ω 3 nh∆∞ng v·∫´n cao h∆°n ƒë√°ng k·ªÉ so v·ªõi Qu√Ω 1. Nh√¨n chung, xu h∆∞·ªõng cho th·∫•y doanh thu tƒÉng m·∫°nh n·ª≠a ƒë·∫ßu nƒÉm, ƒë·∫°t ƒë·ªânh ·ªü gi·ªØa nƒÉm r·ªìi suy gi·∫£m nh·∫π v·ªÅ cu·ªëi nƒÉm.


```python
# Join order_detail v·ªõi order ƒë·ªÉ l·∫•y order_time
order_detail_with_time = (
    dfs["order_detail"]
    .join(dfs["order"].select("order_id", "order_time"), on="order_id", how="inner")
)

# ========================
# Doanh thu theo th√°ng
# ========================
revenue_month = (
    order_detail_with_time
    .withColumn("month", F.month("order_time"))
    .groupBy("month")
    .agg((F.sum(F.col("quantity") * F.col("price"))).alias("total_revenue"))
    .orderBy("month")
    .toPandas()
)

plt.figure(figsize=(10, 5))
sns.lineplot(data=revenue_month, x="month", y="total_revenue", marker="o")
plt.title("üìÖ Doanh thu theo th√°ng")
plt.xlabel("Th√°ng")
plt.ylabel("Doanh thu")
plt.xticks(range(1, 13))
plt.show()

```

Doanh thu theo th√°ng nƒÉm 2024 cho th·∫•y s·ª± ph√¢n h√≥a r√µ r·ªát. Trong 4 th√°ng ƒë·∫ßu nƒÉm, doanh thu duy tr√¨ ·ªü m·ª©c th·∫•p d∆∞·ªõi 2,5 t·ª∑ ƒë·ªìng. T·ª´ th√°ng 5, doanh thu tƒÉng ƒë·ªôt bi·∫øn v√† gi·ªØ ·ªü m·ª©c cao, ƒë·∫°t ƒë·ªânh h∆°n 5,3 t·ª∑ v√†o th√°ng 8. Sau ƒë√≥, t·ª´ th√°ng 9 tr·ªü ƒëi, doanh thu gi·∫£m xu·ªëng c√≤n quanh m·ª©c 3 ‚Äì 3,5 t·ª∑ ƒë·ªìng. ƒêi·ªÅu n√†y ph·∫£n √°nh xu h∆∞·ªõng cao ƒëi·ªÉm v√†o gi·ªØa nƒÉm, tr∆∞·ªõc khi h·∫° nhi·ªát ·ªü giai ƒëo·∫°n cu·ªëi nƒÉm.

```python
# Group tr·ª±c ti·∫øp t·ª´ order_detail
dish_sales = (
    dfs["order_detail"]
    .groupBy("dish_name")
    .agg(F.sum("quantity").alias("total_quantity"))
    .orderBy(F.desc("total_quantity"))
    .toPandas()
)

# V·∫Ω bar chart ngang
plt.figure(figsize=(10, 6))
sns.barplot(
    data=dish_sales,
    y="dish_name",
    x="total_quantity",
    palette="viridis"
)
plt.title("üçΩÔ∏è S·ªë l∆∞·ª£ng b√°n theo m√≥n (to√†n b·ªô d·ªØ li·ªáu 1 nƒÉm)")
plt.xlabel("T·ªïng s·ªë l∆∞·ª£ng b√°n")
plt.ylabel("M√≥n ƒÉn")
plt.show()
```

Trong nƒÉm 2024, c√°c m√≥n ƒÉn c√≥ doanh s·ªë kh√° ƒë·ªìng ƒë·ªÅu, nh∆∞ng n·ªïi b·∫≠t nh·∫•t l√† B√∫n Ch·∫£, H·ªß Ti·∫øu Nam Vang, L·∫©u H·∫£i S·∫£n, M√¨ Qu·∫£ng v√† Ph·ªü B√≤, ƒë·ªÅu ƒë·∫°t tr√™n 40.000 su·∫•t, ch·ª©ng t·ªè s·ª©c h√∫t cao v√† ·ªïn ƒë·ªãnh. Nh√≥m m√≥n c√≤n l·∫°i nh∆∞ G·ªèi Cu·ªën, C∆°m T·∫•m, B√°nh X√®o, B√°nh M√¨ Th·ªãt v√† Ch·∫£ C√° L√£ V·ªçng c√≥ l∆∞·ª£ng b√°n th·∫•p h∆°n, dao ƒë·ªông kho·∫£ng 28.000 ‚Äì 32.000 su·∫•t. ƒêi·ªÅu n√†y cho th·∫•y nh√≥m m√≥n truy·ªÅn th·ªëng v√† ph·ªï bi·∫øn ·ªü c·∫£ ba mi·ªÅn c√≥ l·ª£i th·∫ø l·ªõn v·ªÅ s·ª©c ti√™u th·ª•, trong khi c√°c m√≥n ƒë·∫∑c th√π v√πng mi·ªÅn c√≥ s·ª©c h√∫t h·∫πp h∆°n.

```python
# T√≠nh doanh thu m·ªói m√≥n = SUM(quantity * price)
dish_revenue = (
    dfs["order_detail"]
    .withColumn("revenue", F.col("quantity") * F.col("price"))
    .groupBy("dish_name")
    .agg(F.sum("revenue").alias("total_revenue"))
    .orderBy(F.desc("total_revenue"))
    .toPandas()
)


# V·∫Ω bar chart ngang
plt.figure(figsize=(10, 6))
sns.barplot(
    data=dish_revenue,
    y="dish_name",
    x="total_revenue",
    palette="coolwarm"
)
plt.title("üí∞ Doanh thu theo m√≥n (to√†n b·ªô d·ªØ li·ªáu 1 nƒÉm)")
plt.xlabel("T·ªïng doanh thu")
plt.ylabel("M√≥n ƒÉn")
plt.show()
```

Trong nƒÉm 2024, L·∫©u H·∫£i S·∫£n d·∫´n ƒë·∫ßu v·ªÅ doanh thu, v∆∞·ª£t xa c√°c m√≥n kh√°c v·ªõi g·∫ßn 8 t·ª∑ ƒë·ªìng, d√π s·ªë l∆∞·ª£ng b√°n ra ch·ªâ ngang nh√≥m cao (~40.000 su·∫•t). ƒêi·ªÅu n√†y cho th·∫•y gi√° b√°n trung b√¨nh cao n√™n t·∫°o ra doanh thu v∆∞·ª£t tr·ªôi. T∆∞∆°ng t·ª±, Ch·∫£ C√° L√£ V·ªçng v√† H·ªß Ti·∫øu Nam Vang c≈©ng c√≥ doanh thu n·∫±m trong nh√≥m d·∫´n ƒë·∫ßu, ph·∫£n √°nh m·ª©c gi√° b√°n kh√° cao.

Ng∆∞·ª£c l·∫°i, m·ªôt s·ªë m√≥n nh∆∞ B√∫n Ch·∫£, Ph·ªü B√≤ v√† M√¨ Qu·∫£ng n·∫±m trong nh√≥m c√≥ s·ªë l∆∞·ª£ng b√°n l·ªõn nh·∫•t (h∆°n 40.000 su·∫•t) nh∆∞ng doanh thu ch·ªâ ·ªü m·ª©c trung b√¨nh, cho th·∫•y gi√° b√°n m·ªói su·∫•t th·∫•p h∆°n so v·ªõi c√°c m√≥n l·∫©u hay ƒë·∫∑c s·∫£n. Trong khi ƒë√≥, c√°c m√≥n nh∆∞ C∆°m T·∫•m, B√°nh X√®o, G·ªèi Cu·ªën v√† B√°nh M√¨ Th·ªãt v·ª´a c√≥ s·ªë l∆∞·ª£ng b√°n th·∫•p h∆°n v·ª´a ghi nh·∫≠n doanh thu th·∫•p nh·∫•t, th·ªÉ hi·ªán v·ªã th·∫ø h·∫°n ch·∫ø c·∫£ v·ªÅ s·ª©c ti√™u th·ª• l·∫´n gi√° tr·ªã ƒë∆°n h√†ng.

Nh√¨n chung, d·ªØ li·ªáu cho th·∫•y c√≥ s·ª± kh√°c bi·ªát ƒë√°ng k·ªÉ gi·ªØa m√≥n ƒë∆∞·ª£c chu·ªông v·ªÅ s·ªë l∆∞·ª£ng v√† m√≥n t·∫°o ra doanh thu cao, trong ƒë√≥ nh√≥m m√≥n ƒë·∫∑c s·∫£n/gi√° tr·ªã cao (L·∫©u, Ch·∫£ C√°, H·ªß Ti·∫øu) ƒë√≥ng vai tr√≤ ch√≠nh trong tƒÉng tr∆∞·ªüng doanh thu, c√≤n nh√≥m m√≥n ph·ªï bi·∫øn (B√∫n Ch·∫£, Ph·ªü, M√¨ Qu·∫£ng) ch·ªß y·∫øu ƒë√≥ng g√≥p v·ªÅ s·∫£n l∆∞·ª£ng.

```python
# B1: Th√™m c·ªôt gi·ªù t·ª´ order_time
orders_with_hour = dfs["order"].withColumn("hour", F.hour("order_time"))

# B2: G√°n khung gi·ªù d·ª±a theo c·ªôt hour
orders_with_slot = orders_with_hour.withColumn(
    "time_slot",
    F.when((F.col("hour") >= 6) & (F.col("hour") < 11), "Morning")
     .when((F.col("hour") >= 11) & (F.col("hour") < 14), "Noon")
     .when((F.col("hour") >= 17) & (F.col("hour") < 21), "Evening")
     .otherwise("Night")
)

# B3: ƒê·∫øm s·ªë l∆∞·ª£ng ƒë∆°n h√†ng theo khung gi·ªù
slot_counts = (
    orders_with_slot.groupBy("time_slot")
    .agg(F.countDistinct("order_id").alias("num_orders"))
    .orderBy("num_orders", ascending=False)
    .toPandas()
)


# B4: V·∫Ω bar chart ngang
plt.figure(figsize=(7, 5))
sns.barplot(data=slot_counts, x="num_orders", y="time_slot", palette="Set2")
plt.title("üìä S·ªë l∆∞·ª£ng ƒë∆°n h√†ng theo khung gi·ªù (1 nƒÉm)")
plt.xlabel("S·ªë ƒë∆°n h√†ng")
plt.ylabel("Khung gi·ªù")
plt.show()

```

Trong nƒÉm 2024, l∆∞·ª£ng ƒë∆°n h√†ng ph√¢n b·ªï theo khung gi·ªù c√≥ s·ª± kh√°c bi·ªát ƒë√°ng k·ªÉ. Khung gi·ªù Night chi·∫øm nhi·ªÅu ƒë∆°n nh·∫•t v·ªõi h∆°n 20.000 ƒë∆°n, cho th·∫•y nhu c·∫ßu ƒÉn u·ªëng/ƒë·∫∑t m√≥n v√†o bu·ªïi t·ªëi mu·ªôn r·∫•t cao. Ti·∫øp theo l√† Evening (kho·∫£ng 16.000 ƒë∆°n), ph·∫£n √°nh th√≥i quen ƒÉn t·ªëi ho·∫∑c t·ª• t·∫≠p b·∫°n b√®. Hai khung gi·ªù ban ng√†y (Morning v√† Noon) c√≥ s·ªë ƒë∆°n th·∫•p h∆°n, ch·ªâ quanh 12.000 ƒë∆°n m·ªói khung, ph√π h·ª£p v·ªõi th·ª±c t·∫ø nhu c·∫ßu ƒÉn s√°ng v√† tr∆∞a th∆∞·ªùng ƒë∆°n gi·∫£n, √≠t ph√°t sinh nhi·ªÅu ƒë∆°n h√†ng.

Nh√¨n chung, d·ªØ li·ªáu cho th·∫•y bu·ªïi t·ªëi v√† ƒë√™m l√† th·ªùi ƒëi·ªÉm v√†ng cho doanh thu nh√† h√†ng, chi·∫øm ∆∞u th·∫ø c·∫£ v·ªÅ s·ªë l∆∞·ª£ng ƒë∆°n h√†ng so v·ªõi ban ng√†y.
